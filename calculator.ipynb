```python
from tkinter import *
import tkinter.messagebox
root=Tk()
root.geometry("300x600")
root.title("CALCULATOR")
root.iconbitmap(r"C:\Users\asd\Downloads\calculator.ico")
root.resizable(False,False)
input=""
def show(value):
    global input
    input+=value
    entr.config(text=input)
def clear():
    global input
    input=""
    entr.config(text=input)
def calculate():
    global input
    result=""
    if input!="":
        try:
            result=eval(input)
        except:
            tkinter.messagebox.showwarning(title='Warning',message='Please put an input first')
    entr.config(text=result)
entr=Label(root,textvar=input,font="callibre 25 bold")
entr.pack(pady=4)

f1=Frame(root,bg="grey")
b1=Button(f1,text="9",font="callibre 30 italic",bg="black",fg="white",command=lambda:show("9"))
b1.pack(side=LEFT,padx=5,pady=6)
b2=Button(f1,text="8",font="callibre 30 italic",bg="black",fg="white",command=lambda:show("8"))
b2.pack(side=LEFT,padx=5,pady=6)
b3=Button(f1,text="7",font="callibre 30 italic",bg="black",fg="white",command=lambda:show("7"))
b3.pack(side=LEFT,padx=5,pady=6)
f1.pack()
f2=Frame(root,bg="grey")
b1=Button(f2,text="6",font="callibre 30 italic",bg="black",fg="white",command=lambda:show("6"))
b1.pack(side=LEFT,padx=5,pady=6)
b2=Button(f2,text="5",font="callibre 30 italic",bg="black",fg="white",command=lambda:show("5"))
b2.pack(side=LEFT,padx=5,pady=6)
b3=Button(f2,text="4",font="callibre 30 italic",bg="black",fg="white",command=lambda:show("4"))
b3.pack(side=LEFT,padx=5,pady=6)
f2.pack()
f3=Frame(root,bg="grey")
b1=Button(f3,text="3",font="callibre 30 italic",bg="black",fg="white",command=lambda:show("3"))
b1.pack(side=LEFT,padx=5,pady=6)
b2=Button(f3,text="2",font="callibre 30 italic",bg="black",fg="white",command=lambda:show("2"))
b2.pack(side=LEFT,padx=5,pady=6)
b3=Button(f3,text="1",font="callibre 30 italic",bg="black",fg="white",command=lambda:show("1"))
b3.pack(side=LEFT,padx=5,pady=6)
f3.pack()
f4=Frame(root,bg="grey")
b1=Button(f4,text="+",font="callibre 32 italic",bg="black",fg="white",command=lambda:show("+"))
b1.pack(side=LEFT,padx=5,pady=6)
b2=Button(f4,text="0",font="callibre 30 italic",bg="black",fg="white",command=lambda:show("0"))
b2.pack(side=LEFT,padx=5,pady=6)
b3=Button(f4,text="-",font="callibre 31 italic",bg="black",fg="white",command=lambda:show("-"))
b3.pack(side=LEFT,padx=5,pady=6)
f4.pack()
f5=Frame(root,bg="grey")
b1=Button(f5,text="*",font="callibre 32 italic",bg="black",fg="white",command=lambda:show("*"))
b1.pack(side=LEFT,padx=5,pady=6)
b2=Button(f5,text="/",font="callibre 32 italic",bg="black",fg="white",command=lambda:show("/"))
b2.pack(side=LEFT,padx=5,pady=6)
b3=Button(f5,text="=",font="callibre 32 italic",bg="orange",fg="white",command=lambda:calculate())
b3.pack(side=LEFT,padx=6,pady=10)
f5.pack()
f6=Frame(root,bg="grey")
b1=Button(f6,text="c",font="callibre 30 italic",bg="blue",fg="white",command=lambda:clear())
b1.pack(side=LEFT,padx=5,pady=6)
b3=Button(f6,text=".",font="callibre 30 italic",bg="black",fg="white",command=lambda:show("."))
b3.pack(side=LEFT,padx=5,pady=6)
b2=Button(f6,text="%",font="callibre 30 italic",bg="black",fg="white",command=lambda:show("%"))
b2.pack(side=LEFT,padx=5,pady=6)
f6.pack()
root.mainloop()
```
